```python
import numpy as np
import matplotlib.pyplot as plt

# Time array (in days)
t = np.linspace(0.1, 100, 500)

# Initial pressure
pi = 3000  # psia

# Steady-State
p_steady = np.full_like(t, pi)

# Unsteady-State (transient)
p_unsteady = pi * np.exp(-0.03 * t)

# Pseudo-Steady-State
p_pseudo = pi - 10 * t
p_pseudo[p_pseudo < 0] = 0

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, p_steady, label="Steady-State Flow", linewidth=2)
plt.plot(t, p_unsteady, label="Unsteady-State Flow", linewidth=2)
plt.plot(t, p_pseudo, label="Pseudo-Steady-State Flow", linewidth=2)
plt.title("Pressure vs Time for Different Flow Regimes")
plt.xlabel("Time (days)")
plt.ylabel("Reservoir Pressure (psia)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
